# Connect with Database

```python
import mysql.connector
from getpass import getpass
from datetime import datetime, date, time, timedelta
import pandas as pd
user = input('User name: ') #root
pwd = getpass('Password')

my_db = None

try:
    my_db = mysql.connector.connect (host = 'localhost' , user = user, password = pwd, database = 'jtm_database')
    print('Successfully connected to the database')

except Exception as e:
    print('Something went wrong while attempting to connect the database')
    if my_db is not None: 
        my_db.rollback()
        print(f' >>> {e}')
finally:
    if my_db is not None:
        my_db.rollback()
        print('Database closed')
```

```python
my_pointer = my_db.cursor()

while True:
    i_q = input('Would you like to Insert or Query data? : "i" for Insert or "q" for Query ??')
    if i_q == 'q' or i_q == 'i':
        break
    else:
        print('Please insert only i or q...')
############### INSERT ################
if i_q == 'i':
    while True:
        try:
            memberID = int(input('Insert memberID (Number) with 3 numbers only: '))
            if len(str(memberID)) != 3:
                raise ValueError
            else:
                break
        except ValueError:
            print('Please insert exactly 3 numbers.')
    ####CHECK WHETHER IN memberdetails####
    memde = "select memberID\
        from memberdetails as mem\
        ;"
    my_pointer.execute(memde)
    allmember_lst1 =  my_pointer.fetchall()
    df_1 = pd.DataFrame(allmember_lst1, columns=['Member ID'])
    df_1 = df_1.to_numpy()
    if f"member{memberID}" not in df_1:
        while True: 
            try: 
                name = input('Insert your Name (Characture) :'   )
                if int == int(name):
                    pass
            except:
                break
        while True:
            M_W = input('Are you men or women?? type: "m" for men "w" for women')
            if M_W == 'm' or M_W == 'w' or M_W == 'M' or M_W == 'W':
                break
            else:
                print('Please insert only m or w...')

        while True:
            try:
                weight = int(input('Insert weight(kg.) (Number):'   ))
                break
            except:
                print('Please Insert number :...')

        while True:
            try:
                height = int(input('Insert height(cm.) (Number):'   ))
                break
            except:
                print('Please Insert number :...')
        # insert to database
        insert_to_memberdetails = "INSERT INTO memberdetails (memberID, Name, Gender, Weight, Height) VALUES (%s, %s, %s, %s, %s)"
        val = (f"member{memberID}", f"{name}", f"{M_W}", f"{weight}", f"{height}")
        my_pointer.execute(insert_to_memberdetails, val)
        my_db.commit()
    ####CHECK WHETHER IN fooddetails####
    menuu = "select Menu\
        from fooddetails\
        ;"
    my_pointer.execute(menuu)
    menu_lst1 = my_pointer.fetchall()
    df_2 = pd.DataFrame(menu_lst1, columns=['Menu'])
    df_2 = df_2.to_numpy()
 
    menu = input('Insert your Menu (String): ')

    if menu not in df_2:
        while True:
            try:
                menu_type = input('Insert your Menu type เครื่องดื่ม อาหารคาว ผลไม้ นม ขนม (String): ')
                if menu_type.isalpha() and menu_type in ['เครื่องดื่ม','อาหารคาว','ผลไม้','นม','ขนม']:
                    break
                else:
                    raise ValueError
            except ValueError:
                print('Please insert a string.>>เครื่องดื่ม อาหารคาว ผลไม้ นม ขนม (String)')
        while True:
            try:
                calories = int(input('Insert Calories (Number):'   ))
                break
            except:
                print('Please Insert number of Calories :...')
        # insert to database
        insert_to_fooddetails = "INSERT INTO fooddetails (Menu, Category, calories) VALUES (%s, %s, %s)"
        val = (f"{menu}", f"{menu_type}", f"{calories}")
        my_pointer.execute(insert_to_fooddetails, val)
        my_db.commit()
    ####Collect data to prepare to insert to foodtracker table ####
    while True:
        try:
            channel = input('Insert your channel (Number): 1.Take away 2.vending kiosk 3. delivery 4.dine-in 5.Others ')
            channel = int(channel)
            if 1 <= channel <= 5:
                break
            else:
                raise ValueError
        except ValueError:
            print('Please insert a number between 1 and 5.where 1.Take away 2.vending kiosk 3. delivery 4.dine-in 5.Others')

    #INSERT day mounth year
    while True:
        try:
            year = input('Insert Year ex: 2023 ')
            year = int(year)
            if 1 <= year <= 2023:
                break
            else:
                raise ValueError
        except ValueError:
            print('Please insert a number between 1 and 2023')
    while True:
        try:
            month = input('Insert Month ex: 5 ')
            month = int(month)
            if 1 <= month <= 12:
                break
            else:
                raise ValueError
        except ValueError:
            print('Please insert a number between 1 and 12')
    while True:
        try:
            day = input('Insert Day ex: 31 ')
            day = int(day)
            if 1 <= day <= 31:
                break
            else:
                raise ValueError
        except ValueError:
            print('Please insert a number between 1 and 31')   

    specific_date = date(year, month, day)

    while True:
        try:
            val_meal = {1: 'เช้า',2: 'สาย',3: 'กลางวัน',4: 'บ่าย',5: 'เย็น',6: 'ค่ำ',}
            meal = input('Insert your meal (Number): 1.เช้า 2.สาย 3.กลางวัน 4.บ่าย 5.เย็น 6.ค่ำ ')
            meal = int(meal) 
            if 1 <= meal <= 6:
                break
            else:
                raise ValueError
        except ValueError:
            print('Please insert a number between 1 and 6 where 1.เช้า 2.สาย 3.กลางวัน 4.บ่าย 5.เย็น 6.ค่ำ :')

    while True:
        try:
            shop_name = input('Insert your shop name (String): ')
            if shop_name.isalpha():
                break
            else:
                raise ValueError
        except ValueError:
            print('Please insert a string: ')

    while True:
        try:
            food_price = int(input('Insert price (Number):'   ))
            break
        except:
            print('Please Insert number :...')
    
    # insert to database
    insert_to_foodtracker = "INSERT INTO foodtracker (memberID, date, meal, menu, channel, shopName, price) VALUES (%s, %s, %s, %s, %s,%s, %s)"
    val = (f"member{memberID}", specific_date, val_meal[meal], menu, channel, shop_name, food_price)
    my_pointer.execute(insert_to_foodtracker, val)
    my_db.commit()










################## quiry ################
if i_q == 'q':
    re_ = 'y'
    while re_ == 'y' or re_ == 'Y':
        while True:
            try:
                ans_ = int(input('Insert number 1-5: '))
                if ans_ == 1 or ans_ == 2 or ans_ == 3 or ans_ == 4 or ans_ == 5 :
                    break
                else:
                    raise
            except:
                print('Insert between 1-5')
                
        if ans_ == 1:
                while True:
                    try:
                        print('1.สมาชิกแต่ละคนใช้เงินทั้งหมดไปเท่าไหร่ในการซื้ออาหารที่มีแคลอรี่ต่อชิ้นไม่เกิน xxx แคลอรี่')
                        num = int(input('Insert the maximum calories that you want to calculate: '))
                        query = f"SELECT ft.memberID\
                            , ft.menu, fd.calories\
                            , SUM(price)\
                            FROM foodtracker AS ft\
                            JOIN fooddetails AS fd ON fd.Menu = ft.menu\
                            WHERE fd.calories <= {num}\
                            GROUP BY memberID\
                            , menu\
                            order by memberID;"
                        my_pointer.execute(query)
                        query_result = my_pointer.fetchall()
                        df = pd.DataFrame(query_result, columns=['Member ID', 'Menu', 'Calories per one piece', 'SUM(price)'])
                        display(df)
                        break
                    except ValueError:
                        print('Please insert a valid number.')
        elif ans_ == 2:
            print('2.ประเภทอาหารที่สมาชิกกิน แต่ละชนิดมีราคาทั้งหมด แคลอรี่ทั้งหมด จำนวนการสั่งอาหารประเภทนี้ และแคลอรี่เฉลี่ยต่อจำนวนการสั่ง เท่าไหร่')
            test_2 = f"select memberID\
                ,f.Category\
                ,sum(f.calories) as 'total_calories'\
                ,sum(price) as 'total_price'\
                ,count(*) as 'no_order'\
                ,round(sum(f.calories)/count(*),2) as 'calories_per_order'\
                from foodtracker as trac\
                left join fooddetails as f on trac.menu = f.menu\
                group by memberID\
                ,f.Category\
                order by total_calories desc;"
            my_pointer.execute(test_2)
            lst_2 =  my_pointer.fetchall()
            df = pd.DataFrame(lst_2, columns=['Member ID', 'Category', 'Sum of calories', 'Sum of price', 'Number of order', 'Calories per order'])
            display(df)
        elif ans_ == 3:
            print('3.ในมื้ออาหารแต่ละมื้อสมาชิกแต่ละคนกินไปทั้งหมดกี่แคลอรี่')
            test_3 = f"select ft.memberID\
            ,md.Name\
            ,ft.meal\
            ,sum(fd.calories) as 'total_calories_meal'\
            from foodtracker as ft\
            left join fooddetails as fd ON fd.Menu = ft.menu\
            left join memberdetails as md on md.memberID = ft.memberID\
            group by ft.memberID\
            ,ft.meal\
            order by memberID;"
            my_pointer.execute(test_3)
            lst_3 =  my_pointer.fetchall()
            df = pd.DataFrame(lst_3, columns=['Member ID', 'Name', 'Meal', 'Total calories'])
            display(df)

        elif ans_ == 4:
            print('4.สมาชิกแต่ละคนสั่งอาหารตามช่องทางต่างๆไปจำนวนเท่าไหร่')
            test_4 = f"select  ft.memberID\
                ,ft.channel\
                ,count(channel) as 'total_order_of_member'\
                ,cn.ChannelName\
                from foodtracker as ft\
                left join channel as cn on cn.ChannelID = ft.channel\
                group by memberID \
                ,channel\
                order by memberID ;"
            my_pointer.execute(test_4)
            lst_4 =  my_pointer.fetchall()
            df = pd.DataFrame(lst_4, columns=['Member ID', 'ChannelID', 'Number of order', 'Channel name'])
            display(df)

        elif ans_ == 5:
            print('5.ในช่วงเวลาหนึ่งสมาชิกแต่ละคนสั่งเมนูต่างๆไปจำนวนกี่ครั้ง')
            test_5 = f"select ft.memberID\
                ,md.Name\
                ,ft.menu\
                ,count(menu) as 'total_num_menu'\
                from foodtracker as ft\
                left join memberdetails as md on md.memberID = ft.memberID\
                where ft.date between '2023-01-01' and '2023-12-31'\
                group by ft.memberID\
                ,ft.menu\
                order by memberID ;"
            my_pointer.execute(test_5)
            lst_5 =  my_pointer.fetchall()
            lst_5
            df = pd.DataFrame(lst_5, columns=['Member ID', 'ChannelID', 'Menu', 'Number of menu'])
            display(df)

        re_ = input('Do you want to ask again ? please insert "Y" or Any to quit...')





   
```
